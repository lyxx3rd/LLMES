```python
from sentence_transformers import SentenceTransformer

sentences = ["通过对人力资源进行有效的管理和控制，为公司确定和提供所需的人力资源，保证人力资源的及时引进、针对培训和合理分配，满足公司的需要，使从事各职位的任职人员能够胜任岗位要求，从而确保质量管理体系有效运行所需的人力资源的适宜性和充分性。", 
             "适用于公司内部职责、权限的规定和人力资源的配置，公司所有对产品质量有影响的工作人员及操作人员培训、考核工作。",
             "人力资源管理体系：指以职位分析为起点，以实现“为事择人，任人唯贤、专业对口、事得其人”的人力资源配置目标，并建立相应的薪酬福利体系、绩效考核体系和员工培训体系；"]
model = SentenceTransformer('./yangjhchs/acge_text_embedding')
Index = model.encode(sentences, normalize_embeddings=True)
```

```python
question = ["第二章 机构与人员, 第五条 企业应当建立与医疗器械生产相适应的管理机构。"]
question_embeddings = model.encode(question, normalize_embeddings=True)
similarity = Index @ question_embeddings.T
print(similarity)
```
